Treating Null Values

```python
df.isnull().sum()
```

```python
df['Size'].unique()
```

```python
df.head()
```

```python
df.groupby('Category')["Size"].unique()
```

```python
df.groupby('Item Purchased')["Size"].unique()
```

```python
# Treating null values replacing values in Electronics Category

df.loc[df["Category"]=='Electronics','Size']='Not Applicable'
```

```python
df.groupby('Item Purchased')["Size"].unique()#now size is changed succssfully for that column size is not applicable
```

```python
# Treating null values replacing values in Accessories Category

df.loc[df["Category"]=='Accessories','Size']='Not Applicable'
```

```python
df.groupby('Category')["Size"].unique() # accessories size treatment
```

```python
# Treating null values replacing values in footwear Category

df.loc[df["Category"]=='Footwear','Size']='Not Applicable'
```

```python
df.groupby('Category')["Size"].unique()#Size tratment for footwear
```

```python
df.loc[(df["Category"]=="Clothing")& (df["Size"].isnull()),"Size"]=df[df["Category"]=="Clothing"]["Size"].mode()[0]
```

```python
df.groupby('Category')["Size"].unique()#Null Values Treatment
```

```python
#Mean of review rating
df["Review Rating"].mean()
```

```python
#Median of review rating
df["Review Rating"].median()
```

```python
#Mode of review rating
df["Review Rating"].mode()
```

```python
df.groupby("Item Purchased")["Review Rating"].mean()
```

```python
#filling null values of Review Ratings with he mean value
df["Review Rating"]=df["Review Rating"].fillna(df.groupby("Item Purchased")["Review Rating"].transform("mean"))
```

```python
df.isnull().sum()
```

```python
df["Review Rating"]=df["Review Rating"].round(2)# round values 
```

```python
df.head()
```

```python
#filling Previous purchase with zero(0)
df["Previous Purchases"]=df["Previous Purchases"].fillna(0)
```

```python
df.isnull().sum()
```

Purchase Amount(USD)

```python
df.groupby("Item Purchased")["Purchase Amount (USD)"].mean()
```

```python
#filling null values of Purchase Amount(USD) with he mean value
df["Purchase Amount (USD)"]=df["Purchase Amount (USD)"].fillna(df.groupby("Item Purchased")["Purchase Amount (USD)"].transform("mean"))
```

```python
df.isnull().sum()
```
